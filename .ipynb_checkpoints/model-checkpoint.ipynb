# Creating a model for audio features
### Importing dataframes with audio features of hot 100 and spotify playlist

### Filtering only the columns needed for training the model

```python
audio_feat = audio_feat_all[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']]
#Removed duration ms to see what changes
```

```python
audio_feat
```

### Scaling the dataframe

```python
scaler = StandardScaler()
scaler.fit(audio_feat)
audio_scaled = scaler.transform(audio_feat)
audio_feat_scaled = pd.DataFrame(audio_scaled, columns = audio_feat.columns)

#Not scaled
display(audio_feat.head()) 
#Scaled
display(audio_feat_scaled.head())
```

### Finding optimal amount of k clusters

```python
#Deelete later not needed just to check
kmeans2 = KMeans(n_clusters=8,
                init="random",
                n_init=3,  
                max_iter=6,
                tol=0,
                algorithm="full",
                random_state=1234)
kmeans2.fit(audio_feat_scaled)

print(kmeans2.inertia_)
```

```python
K = range(2, 21)
inertia = []

for k in K:
    print("Training a K-Means model with {} neighbours! ".format(k))
    print()
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(audio_feat_scaled)
    inertia.append(kmeans.inertia_)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
K = range(2, 20)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(audio_feat_scaled)
    
    #filename = "/content/drive/MyDrive/kmeans_" + str(k) + ".pickle"
    #with open(filename, "wb") as f:
        #pickle.dump(kmeans,f)
        
    silhouette.append(silhouette_score(audio_feat_scaled, kmeans.predict(audio_feat_scaled)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Silhouette Method showing the optimal k')
```

### Retrieving the clusters out of the model

```python
kmeans = KMeans(n_clusters=7, random_state=1234)
kmeans.fit(audio_feat_scaled)
```

```python
clusters = kmeans.predict(audio_feat_scaled)

pd.Series(clusters).value_counts().sort_index()
```

```python
audio_feat_all["cluster"] = clusters
audio_feat_all.head()
```
