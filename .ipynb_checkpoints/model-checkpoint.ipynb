# Hot Music Recommender
## Music recommender model
Creating a music recommender by using the Sklearn K-means algorithm with information about songs audio features. The audio features will be used to create clusters of similar sounding songs in the model. This way the music recommendation is hopefully liked by the user, given the fact that the recommendation is based on the users input.
- https://medium.com/analytics-vidhya/practical-approach-to-kmeans-clustering-python-and-why-scaling-is-important-44ac0b0fea47

#### Merging both dataframes


```python
audio_feat_all = pd.concat([spotify_songs, audiofeat_hot100])
```

#### Optimal columns
Checking out all the columns of the new, merged dataframe and thinking about which ones to use for training the model. Training the model with only the necessary information not only improves its performance, but also saves computing and running time!

```python
audio_feat_all.columns
```

#### Preparing dataframe for optimal modeling
Removing columns that are irrelevant or might even decrease the models performance by applying a filter and saving the model dataframe in audio_feat.  
- Removed "duration_ms" since the lenght of a song is not a good characteristic to use for clustering similar sounding songs  
Since all columns contain numerical values, I don't need to transform any categorical values :)

```python
audio_feat = audio_feat_all[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']]
```

```python
audio_feat.head()
```

#### Scaling the dataframe
Scaling the dataframe puts all the values in it into relation with each other. First assigning the scaler to a variable and then fitting it on the dataframe
Below I printed both dataframes, the unscaled and the scaled one. In the column key you can see the change after scaling the values very nicely.

```python
scaler = StandardScaler()

scaler.fit(audio_feat)

audio_scaled = scaler.transform(audio_feat)

audio_feat_scaled = pd.DataFrame(audio_scaled, columns = audio_feat.columns)
```

```python
#Not scaled
display(audio_feat.head()) 
#Scaled
display(audio_feat_scaled.head())
```

#### Finding optimal amount of clusters
which is also called k, using 2 different methods:
1. Elbow Method
2. Silhouette Method

```python
#Delete later not needed just to check
kmeans2 = KMeans(n_clusters=8,
                init="random",
                n_init=3,  
                max_iter=6,
                tol=0,
                algorithm="full",
                random_state=1234)
kmeans2.fit(audio_feat_scaled)

print(kmeans2.inertia_)
```

#### Elbow Method

```python
K = range(2, 21)
inertia = []

for k in K:
    print("Training a K-Means model with {} neighbours! ".format(k))
    print()
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(audio_feat_scaled)
    inertia.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

#### Silhouette Method

```python
K = range(2, 20)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(audio_feat_scaled)
        
    silhouette.append(silhouette_score(audio_feat_scaled, kmeans.predict(audio_feat_scaled)))
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Silhouette Method showing the optimal k')
```

#### Saving the clusters 
Fitting the K-means on the audio features dataframe and then predicting how many songs belong to which cluster, which is then turned into a column and added to the audio features dataframe.

```python
kmeans = KMeans(n_clusters=7, random_state=1234)
kmeans.fit(audio_feat_scaled)
```

```python
clusters = kmeans.predict(audio_feat_scaled)

pd.Series(clusters).value_counts().sort_index()
```

```python
audio_feat_all["cluster"] = clusters
audio_feat_all.head()
```

Using a filter to create a final dataframe only containing the needed columns, as well as filling any null values.

```python
audio_feat_all = audio_feat_all[['artist', 'song', 'id', 'danceability', 'energy', 'key','loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness','liveness', 'valence', 'tempo', 'duration_ms', 'time_signature','origin', 'cluster']]
```

```python
audio_feat_all = audio_feat_all.fillna(1)
```
